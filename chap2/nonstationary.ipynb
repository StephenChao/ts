So far we have focused on return series that are stationary. In some studies, interest rates, foreign exchange rates, or the price series of an asset are of interest. These series tend to be nonstationary. For a price series, the nonstationarity is mainly due to the fact that there is no fixed level for the price. In the time series literature, such a nonstationary series is called unit-root nonstationary time series. The best known example of unit-root nonstationary time series is the random-walk model.


### Simple random walk

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf

# ----------------------
# 1. 生成随机游走数据
# ----------------------
np.random.seed(42)  # 固定随机种子
n = 500             # 数据点数量
s0 = 100            # 初始股价
mu = 0.0            # 漂移项（通常假设为0）
sigma = 1.0         # 波动率（标准差）

# 生成随机步长（每日价格变化）
steps = np.random.normal(mu, sigma, n)

# 计算累积价格（随机游走）
price = s0 + np.cumsum(steps)

# 转换为时间序列（Pandas Series）
dates = pd.date_range(start="2020-01-01", periods=n, freq="D")
price_series = pd.Series(price, index=dates, name="Price")

# ----------------------
# 2. 绘制价格走势图
# ----------------------
plt.figure(figsize=(12, 6))
plt.plot(price_series, label="Random Walk Price", color="blue")
plt.title("Simulated Stock Price (Random Walk)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# ----------------------
# 3. 绘制ACF（自相关函数）
# ----------------------
plt.figure(figsize=(12, 6))
plot_acf(price_series, lags=40, alpha=0.05, title="ACF of Random Walk Price")
plt.xlabel("Lags")
plt.ylabel("Autocorrelation")
plt.grid(alpha=0.3)
plt.show()
```
